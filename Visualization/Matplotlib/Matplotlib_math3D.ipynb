```python
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-10, 10, 0.23)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)

def show(X, Y, Z):
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
    plt.show()
```

```python
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

show(X, Y, Z)
```

```python
u = np.exp(X)*np.cos(Y)

Z = u

show(X, Y, Z)
```

```python
u = np.log(np.abs(np.sqrt(X**2 + Y**2)))

Z = u

show(X, Y, Z)
```

```python
u = np.arctan(Y/X)

Z = u

show(X, Y, Z)
```

```python
u = np.exp(2*np.log(np.abs(np.sqrt(X**2 + Y**2))))

Z = u

show(X, Y, Z)
```

```python
Z = np.sin(X**2+Y**2)

show(X, Y, Z)
```
